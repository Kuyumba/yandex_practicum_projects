#### 2.2.4. Cтолбец `rating` 

```python
# дата основания ESRB - 1994 год.
# Посмотрим, как это могло повлиять на пропуски в столбце с рейтингом ESRB
data_under1994 = data.query('year_release <= 1994')
print('Число игр старше 1994 года: ', data_under1994['rating'].shape[0])
print('Число пропусков в rating среди игр старше 1994 года: ', data_under1994['rating'].isnull().sum())
print('Число пропусков в user_score среди игр старше 1994 года: ', data_under1994['user_score'].isnull().sum())

print()
print('Строки без пропусков в рейтинге для игр старше 1994 года: ')
display(data_under1994[data_under1994['rating'].notnull()])
```

ESRB был основан в 1994 году. Из 505 игр, выпущенных ранее 1994 года, оценку получили лишь 5. Видимо, это те, которые сохранили популярность до 1994 года. То есть оценок ESRB до 1994 года почти нет. \
Но это лишь малая часть пропусков в столбце`rating` - 500 из 6766 (7%).

Пропуски в столбцах с оценками и рейтингами - масовые, и часто синхронные. Часть этих пропусков связана с возрастом игр - эти оценки начали проводить, когда многие игры уже перестали продаваться и использоваться (как мы только что видели на примере рейтинга ESRB для игр, выпущенных ранее 1994 года). Восстановить эти оценки не представляется возможным, тем более что во многих случаев этих оценок никогда и не было. Заменять их какими-либо универсальными значениями (нулями или средними значениями) тоже нецелесообразно: это может исказить нам дальнейшие статистические оценки. Поэтому, не будем пока торопиться с удалением пропусков в столбцах с оценками и рейтингами.

## Шаг 5. Проверка гипотез

    Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
    Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.


```python
print('Числовое описание пользовательских оценок для платформы PC')
var_pc = np.var(pc2012['user_score'], ddof=1)
print('дисперсия: ', var_pc)
print(pc2012['user_score'].describe())
print()
print('Числовое описание пользовательских оценок для платформы Xbox One')
var_xone = np.var(xone['user_score'], ddof=1)
print('дисперсия: ', var_xone)
print(xone['user_score'].describe())
```

Надо поверить гипотезу: средние пользовательские рейтинги платформ XOne и PC одинаовые. 

Будем проверять гипотезу о равенстве средних двух генеральных совокупностей по взятым из них выборкам: \
рейтинги PC и рейтинги XOne - это две разные генеральные совокупности. \
Воспользуемся методом scipy.stats.ttest_ind() \
Нулевая гипотеза всегда формулируется так, чтобы использовать знак равенства. Значит:
- Нулевая гипотеза (Н0): средние пользовательские рейтинги платформ XOne и PC одинаовые.
- Альтернативная гипотеза (H1): средние пользовательские рейтинги платформ XOne и PC различаются

- Мы не убирали из наших выборок пропуски и поэтому, чтобы не получить `pvalue = nan` и чтобы тест отработал корректно, мы должны либо отфильтровать наши выборки от пропусков, либо использовать параметр `nan_policy` в тесте.


- Дисперсии наших выборок хоть и одного порядка, но отличаются в полтора раза. Назначим `equal_var=False`.


```python
# Дисперсии наших выборок хоть и одного порядка, но отличаются в полтора раза
# Зададим equal_var=False

alpha = 0.05
results = st.ttest_ind(pc2012['user_score'], xone['user_score'], equal_var=False, nan_policy='omit')
print('p-значение: ', results.pvalue)
if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

```python
### Мы не удаляли пропуски из выборки:

pc2012[pc2012['user_score'].isnull() == True].head()
```

Попробуем другой способ  -  перед проверкой гипотез удалим пропуски.

```python
xone_drop = xone.dropna(subset = ['user_score'])
pc2012_drop = pc2012.dropna(subset = ['user_score'])
#display(xone_drop.head())
#display(pc2012_drop.head())
alpha = 0.05
results = st.ttest_ind(pc2012_drop['user_score'], xone_drop['user_score'], equal_var=False)
print('p-значение: ', results.pvalue)
if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

- Результат в обоих случаях почти одинаовый. p-value = 0.549.

- Такой результат не даёт нам основания отвергнуть гипотезу о том, что средние пользовательские рейтинги платформ XOne и PC одинаовые.

- При условии, что нулевая гипотеза верна, вероятность получить наблюдаемые в нашей выборке результаты составляет 55%.

Будем проверять гипотезу о равенстве средних двух генеральных совокупностей по взятым из них выборкам:\
рейтинги Action и рейтинги Sports - это две разные генеральные совокупности.\
Воспользуемся методом scipy.stats.ttest_ind()\
Нулевая гипотеза всегда формулируется так, чтобы использовать знак равенства. Значит:

- Нулевая гипотеза (Н0): средние пользовательские рейтинги жанров Action и Sports одинаовые.
- Альтернативная гипотеза (H1): средние пользовательские рейтинги жанров Action и Sports различаются


```python
# Подготовим выборки. удалим пропуски и сформируем выборки по жанрам.
data2012_drop = data2012.dropna(subset = ['user_score'])
action_drop = data2012_drop.query('genre == "Action"')
sports_drop = data2012_drop.query('genre == "Sports"')

print('Числовое описание пользовательских оценок для жанра Action')
var_action = np.var(action_drop['user_score'], ddof=1)
print('дисперсия: ', var_action)
#print(action_drop['user_score'].describe())
print()
print('Числовое описание пользовательских оценок для жанра Sports')
var_sports = np.var(sports_drop['user_score'], ddof=1)
print('дисперсия: ', var_sports)
#print(sports_drop['user_score'].describe())
#display(action_drop.head())
#display(sports_drop.head())
```

```python
alpha = 0.05
results = st.ttest_ind(action_drop['user_score'], sports_drop['user_score'], equal_var=False)
print('p-значение: ', results.pvalue)
if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```
